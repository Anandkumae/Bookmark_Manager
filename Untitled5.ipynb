```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("/content/drive/MyDrive/Titanic (1).csv")
data.head()
```

```python
data.info()
```

```python
mean_age=data.age.mean()
print(mean_age)
mean_fare=data.fare.mean()
mean_fare
```

```python
data.age.fillna(mean_age,inplace=True)
data.fare.fillna(mean_fare,inplace=True)
```

```python
data.info()
```

```python
data1=pd.get_dummies(data)
data1
```

```python
data1.info()
```

```python
#step2-identify input and output
x=data1.drop(['survived'],axis=1)
y=data1['survived']
print(x.head(2))
print(y.head(2))
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
print(x_train.head())
print(y_train.head())
print(x_test.head())
print(y_test.head())
```

```python
#fitting
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_test_pred=model.predict(x_test)
y_test_pred
```

```python
from sklearn.naive_bayes import GaussianNB
model1=GaussianNB()
model1.fit(x_train,y_train)
```

```python
y_test_pred_naive=model1.predict(x_test)
y_test_pred_naive
```

```python
y_test_pred_proba=model1.predict_proba(x_test)
y_test_pred_proba
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_test_pred_naive))
print(accuracy_score(y_test,y_test_pred))
```
